#### 14. Write a python program to scrap 10 images of Rayban Sunglasses from flipkart website and save them in a directory.

```python
#Importing libraries
import requests
from requests import get
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
import shutil
from random import choice
from string import ascii_uppercase
```

```python
#Defining parameters
ScrapUrl = "https://www.flipkart.com/search?q=rayban+sunglasses+for+men&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&as-pos=2&as-type=RECENT&suggestionId=rayban+sunglasses+for+men%7CSunglasses&requestId=ce9ce1d4-65d7-45cc-bff5-04d8e1e3837f&as-backfill=on"
```

```python
#Creating download function to download images            
def DownloadImage(Image):
    Response = requests.get(Image[0], stream=True)
    ImageName  = ''.join(choice(ascii_uppercase) for i in range(15))
    
    file = open("webscrapping/rayban/{}.jpg".format(ImageName), 'wb')
    Response.raw.decode_content = True
    shutil.copyfileobj(Response.raw, file)
    del Response
```

```python
#Adding chrome driver
ChromeDriver = webdriver.Chrome("C:/Python/Python37/chrome/chromedriver.exe")
ChromeDriver.get(ScrapUrl)
Content = ChromeDriver.page_source
Soup = BeautifulSoup(Content, "html.parser")

#Initiate data storage
ImageInfo = []
ImageReq = 20
ImageCount = 0

#looping multiple pages
for PageVisits in range(1,100):

    GlassDiv = Soup.find_all('div', class_='_3ZJShS _31bMyl')

    #Loop through each Container
    for Container in GlassDiv:

        #Checking for the image count
        if(ImageCount < ImageReq):
            
            image_tag = Container.findChildren('img')
            ImageInfo.append((image_tag[0]["src"], image_tag[0]["alt"]))
            ImageCount = ImageCount+1

        else:
            break

    if(ImageCount < ImageReq):

        #Getting number of buttons i.e. Next and Previous
        ButtonCount  = len(ChromeDriver.find_elements_by_xpath("//a[@class='_3fVaIS']"))

        #If there are two button with same class, it will click on Next button and if only single button it will click on Next only
        if ButtonCount > 1:
            SubmitButton = ChromeDriver.find_elements_by_xpath("//a[@class='_3fVaIS']")[1]
            SubmitButton.click()
        else:
            SubmitButton = ChromeDriver.find_elements_by_xpath("//a[@class='_3fVaIS']")[0]
            SubmitButton.click()
    else:
        break

    #Initializing sleep of 2 second
    sleep(2)

    
for i in range(0, len(ImageInfo)):
    DownloadImage(ImageInfo[i])
```

#### 15 Write a program to scrap 20 mouses’ data which includes model name, price from www.amazon.in whose price is less than 500 and make a data frame with 2 columns “model_name”,” price” with the scraped data.

```python
#Defining parameters
ScrapUrl = "https://www.amazon.in/s?k=laptop+mouse&crid=1PJ1CV0CUIWO3&sprefix=laptop+mouse%2Caps%2C373&ref=nb_sb_ss_i_1_12"
```

```python
#Adding chrome driver
ChromeDriver = webdriver.Chrome("C:/Python/Python37/chrome/chromedriver.exe")
ChromeDriver.get(ScrapUrl)
Content = ChromeDriver.page_source
Soup = BeautifulSoup(Content, "html.parser")


#Initiate data storage
ModelInfo = []
ModelReq = 20
ModelCount = 0

#looping multiple pages
for PageVisits in range(1,100):

    ModelName = Soup.find_all('div', class_='sg-col-inner')

    #Loop through each Container
    for Container in ModelName:

        #Checking for the image count
        if(ModelCount < ModelReq):
            
            ModelName = Container.find('span',attrs={'class':'a-size-medium a-color-base a-text-normal'})
            ModelPrice = Container.find('span', attrs={'class':'a-price-whole'})
            
            if(ModelPrice is not None):
                ModelPrice = ModelPrice.text
                ModelPrice = ModelPrice.replace(',','')
                
                if(int(ModelPrice) <= 500):
                    
                    if(ModelName is not None):
                        ModelName = ModelName.text
                        ModelInfo.append((ModelName, ModelPrice))
                        
                        ModelCount = ModelCount+1
                    

        else:
            break

```

```python
import pandas as pd

df = pd.DataFrame(ModelInfo,columns=['Model Name','Price'])

df
```
